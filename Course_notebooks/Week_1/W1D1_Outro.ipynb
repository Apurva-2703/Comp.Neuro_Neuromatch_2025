# PART 1: Logic of modeling

Models help answer three types of questions about the brain:

* Descriptive - What is the best description of the behavior of this system?
* Mechanistic - How do different mechanisms make this system behave this way?
* Interpretive - Why do neurons' firing rates behave the way they do?

In our case, the exploration on the spikes dataset told us that:

* Descriptively = An exponential decay model best explains the behavior of neuronal spiking and ISI.
* Mechanistically = Excitatory/inhibitory inputs and variable changes in membrane potential explain how the spike activity of the neuron is distributed, as illustrated by the LIF model.
* Intepretively = The exponentially-distributed spiking maximizes information transmission due optimization of entropy.

## Marr's 3 levels of analysis: Hierarchy of complexities.

Computational level - 1
* What is the objective of the system?
* How close is it to optimal?

Algoritmic level - 2
* What are the data structures
* What are the approximations
* What is the runtime?

Implementation level - 3
* What is the hardware?

# PART 3: The flavors of modeling

Models differ in generality, scale, abstraction, and goals.

Knowing models are placed in this space is crucial:

* Define model goals (These can be variable: behavioral/microscopic/macroscopic/behavioral realism, clinical models, inspire experiments, analytically tractable)
* Define evaluation framwork
* What can and should be expected from models
* Determine limitations
