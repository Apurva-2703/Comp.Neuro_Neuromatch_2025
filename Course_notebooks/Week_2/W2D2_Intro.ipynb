## What is a Dynamical System?

A **dynamical system** is defined by:

- **Variables**: Quantities describing the system (e.g., neural activity, connectivity).
- **State space**: The set of all possible states the system can be in  
  - Can be **discrete** or **continuous**.
- **Update rules**: How the state evolves over time (often an **ODE**—ordinary differential equation).

**Intuition:**  
A dynamical system is simply a *rule for how your system changes over time.*

---

## Key Neuroscience Example

**Hodgkin–Huxley model (1952)**  
A landmark model describing spike generation using coupled differential equations.  
Nobel Prize–winning work demonstrating how membrane voltage dynamics arise from ion channel conductances.

---

## Linear Dynamical Systems (LDS)

In linear systems, the future state is a **linear combination** of current activity and connectivity.

### Applications in Neuroscience

- **Oculomotor control**  
  Stability of eye movements can be explained using eigenvalues (which determine system timescales).

- **Decision-making**  
  The *leaky competing accumulator model* exhibits dynamics similar to the  
  **Sequential Probability Ratio Test (SPRT)**.

---

## Eigenvalues & Eigenvectors

- **Eigenvalues** determine system **timescales**: how fast activity decays or grows.
- **Eigenvectors** determine **directions in state space** along which dynamics unfold.

If eigenvectors are **not orthogonal**:

- Integration of inputs depends on their alignment.
- The system’s own dynamics, not just the inputs, decide what information is accumulated.

---

## Transient Dynamics

Even when eigenvalues suggest decay, systems can briefly show **growth** or **slow timescales**.

Related concepts:

- **Non-normality** (Trefethen & Embree)
- **Longer timescales** (Goldman, *Neuron*)
- **Growth despite λ < 0** (Miller, *Neuron*)

---

## Impulse Response

The response of a system to a **brief input**.  
This characterizes the inherent temporal dynamics of the system.

---

## Networks of Linear Systems

Network-level behavior can be decomposed into sums of **simple connectivity motifs**  
(Hu et al., 2018).

---

## Stochastic Linear Dynamical Systems

Add noise to a linear system:

- Covariance structure reveals how information flows through long paths.
- Dimensionality can be inferred from connectivity.
- Leads naturally to **Gaussian process** interpretations of neural dynamics.

---

## High-Level Equation

A generic update rule:
Change in state = Rate of change × Time step

or, in differential form:

\begin{equation}
x_{t+1} = x_t + \dot{x}(t)\, dt
\end{equation}

This is the basis of **Euler integration**, commonly used to simulate neural ODEs.
