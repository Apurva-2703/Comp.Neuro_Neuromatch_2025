## What is a dynamical system?

Defined by:

* Variables: the quantities that describe the system (e.g., neural activity, connectivity).

* State space: the collection of all possible states the system can be in.

 * Can be discrete or continuous.

* Update rules: how the state evolves over time, often given by an ODE (ordinary differential equation).

## Key Neuroscience Example

Hodgkin & Huxley model (1952, Nobel Prize): described spike generation with coupled differential equations.

## Linear Dynamical Systems (LDS)

In linear systems, the future state is a sum of influences (activity + connectivity).

Applications:

Oculomotor control: how the brain keeps eyes stable ‚Üí dynamics can be explained by eigenvalues (which set timescales).

Decision-making: leaky competing accumulator model ‚Üí behavior resembles a Sequential Probability Ratio Test (SPRT).

## Eigenvalues & Eigenvectors

Eigenvalues determine timescales (decay or growth).

Eigenvectors define directions of influence in state space.

If eigenvectors are not orthogonal:

* Integration of inputs depends on their alignment.

* System dynamics‚Äînot just inputs‚Äîdecide what info is accumulated.

## Transient Dynamics

Sometimes the system shows growth or slow timescales even when eigenvalues suggest decay.

Related concepts:

* Non-normality (Trefethen & Embree).

* Longer timescales (Goldman, Neuron).

* Growth despite Œª < 0 (Miller, Neuron).

## Impulse Response

Response of a system to a brief input ‚Üí characterizes the system‚Äôs dynamics.

## Networks of Linear Systems

Network dynamics can be decomposed into sums of simple connectivity motifs (Hu et al., 2018).

## Stochastic Linear Dynamical Systems

Add randomness (noise).

Covariance structure reveals how information flows through long paths in the system.

Dimensionality can be measured from connectivity.

Leads to Gaussian process formulations of neural dynamics.
